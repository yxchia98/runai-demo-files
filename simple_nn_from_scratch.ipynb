# Simple Neural Network from Scratch
This notebook demonstrates training a small neural network from scratch using only NumPy.
We will generate synthetic data and train a model to classify it.

```
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## Generate Sample Training Data

```
# Create a simple 2D classification dataset
n_samples = 200

X = np.random.randn(n_samples, 2)
y = (X[:, 0] * X[:, 1] > 0).astype(int).reshape(-1, 1)

# Visualize
plt.scatter(X[:, 0], X[:, 1], c=y.flatten())
plt.title('Synthetic Dataset')
plt.show()
```

## Define Neural Network

```
# Initialize parameters
input_size = 2
hidden_size = 4
output_size = 1

W1 = np.random.randn(input_size, hidden_size) * 0.1
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size) * 0.1
b2 = np.zeros((1, output_size))

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)
```

## Training Loop

```
learning_rate = 0.1
epochs = 1000

losses = []

for epoch in range(epochs):
    # Forward pass
    z1 = X @ W1 + b1
    a1 = sigmoid(z1)
    z2 = a1 @ W2 + b2
    a2 = sigmoid(z2)

    # Loss (binary cross entropy)
    loss = -np.mean(y * np.log(a2 + 1e-8) + (1 - y) * np.log(1 - a2 + 1e-8))
    losses.append(loss)

    # Backward pass
    dz2 = a2 - y
    dW2 = a1.T @ dz2
    db2 = np.sum(dz2, axis=0, keepdims=True)

    da1 = dz2 @ W2.T
    dz1 = da1 * sigmoid_derivative(a1)
    dW1 = X.T @ dz1
    db1 = np.sum(dz1, axis=0, keepdims=True)

    # Update weights
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {loss:.4f}")
```

## Loss Curve

```
plt.plot(losses)
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

## Predictions

```
preds = (a2 > 0.5).astype(int)

plt.scatter(X[:, 0], X[:, 1], c=preds.flatten())
plt.title('Model Predictions')
plt.show()
```
